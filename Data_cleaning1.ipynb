```python
df_nz_admin = df_nz_admin.dropna(how="any", axis=0)  # drop row contain NaN
print(len(df_nz_admin))
df_nz_admin
```

```python
df_classification = df_nz_admin[df_nz_admin["字段5"].str.startswith("classification")]
df_classification[["字段5"]]
```

```python
# classification: Administration & Office SupportAdministration & Office SupportsubClassification: PA, EA & SecretarialPA, EA & Secretarial
# remove str before Office SupportsubClassification:
df_classification["classification"] = df_classification.apply(
    lambda row: row["字段5"].split("Office SupportsubClassification:")[-1].strip(), axis=1
)
df_classification[["字段5", "classification"]]
```

```python
# remove repeat str, split from the midddle
df_classification["classification"] = df_classification.apply(
    lambda row: row["classification"][len(row["classification"]) // 2 :], axis=1
)

df_classification[["classification"]]
```

```python
df_classification = df_classification.drop(columns=["字段5"])
df_classification
```

```python
# filter the time
df_date = df_classification[df_classification["字段4"].str.contains("ago")]  #
df_date
```

```python
from dateutil.relativedelta import relativedelta
import datetime
def get_past_date(row):
    str_days_ago = row['字段4']
    # remove xxxd+ ago
    str_days_ago = str_days_ago.replace('+','')
    TODAY = datetime.datetime.now()
    splitted = str_days_ago.strip().split()[0]
    unit = splitted[-1].lower()
    num = splitted[:-1]
    if unit in ['hour', 'hours', 'hr', 'hrs', 'h']:
        date = datetime.datetime.now() - relativedelta(hours=int(splitted[0]))
        return date
    elif unit in ['day', 'days', 'd']:
        date = TODAY - relativedelta(days=int(splitted[0]))
        return date
    elif unit in ['wk', 'wks', 'week', 'weeks', 'w']:
        date = TODAY - relativedelta(weeks=int(splitted[0]))
        return date
    elif unit in ['mon', 'mons', 'month', 'months', 'm']:
        date = TODAY - relativedelta(months=int(splitted[0]))
        return date
    elif unit in ['yrs', 'yr', 'years', 'year', 'y']:
        date = TODAY - relativedelta(years=int(splitted[0]))
        return date
    else:
        print(str_days_ago)
        return "Wrong Argument format"
```

```python
# convert xxx ago to datetime
df_date["date"] = df_date.apply(get_past_date, axis=1)
# df_date["date"] = pd.to_numeric(df_date["date"])
df_date[["字段4", "date"]].sort_values(by="date", ascending=False)[:10]
```
